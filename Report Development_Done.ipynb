```python
# Jan
df_jan = df_5mo.loc[df_5mo[d[5]] == 'Jan']
print(df_jan.shape)

# Feb
df_feb = df_5mo.loc[df_5mo[d[5]] == 'Feb']
print(df_feb.shape)

#Mar
df_mar = df_5mo.loc[df_5mo[d[5]] == 'Mar']
print(df_mar.shape)

#Apr
df_apr = df_5mo.loc[df_5mo[d[5]] == 'Apr']
print(df_apr.shape)

#May
df_may = df_5mo.loc[df_5mo[d[5]] == 'Mei']
print(df_may.shape)
```

```python
# Jan
df_jan['temp'] = list(zip(df_jan[d[3]],df_jan[d[4]]))
df_jan['temp'] = df_jan['temp'].apply(lambda x: list(x))
df_jan = df_jan.set_index(d[2])
idx_jan = df_jan.index.repeat(df_jan['temp'].str.len(), axis = None)
rep_jan = pd.DataFrame({'Impressions':np.concatenate(df_jan['temp'].values)},index = idx_jan)
rep_jan['Type'] = ['Mono','Colour','Mono','Colour','Mono','Colour','Mono','Colour','Mono','Colour','Mono','Colour']
df_jan = df_jan.join(rep_jan, how='left')
df_jan.drop(columns = ['Mono Pages','Colour Pages','temp'], inplace = True)
```

```python
six = ['Mono','Colour','Mono','Colour','Mono','Colour','Mono','Colour','Mono','Colour','Mono','Colour']
five = ['Mono','Colour','Mono','Colour','Mono','Colour','Mono','Colour','Mono','Colour']
len(five)
```

```python
# Feb
df_feb['temp'] = list(zip(df_feb[d[3]],df_feb[d[4]]))
df_feb['temp'] = df_feb['temp'].apply(lambda x: list(x))
df_feb = df_feb.set_index(d[2])
idx_feb = df_feb.index.repeat(df_feb['temp'].str.len(), axis = None)
rep_feb = pd.DataFrame({'Impressions':np.concatenate(df_feb['temp'].values)},index = idx_feb)
rep_feb['Type'] = six
df_feb = df_feb.join(rep_feb, how='left')
df_feb.drop(columns = ['Mono Pages','Colour Pages','temp'], inplace = True)


# Mar
df_mar['temp'] = list(zip(df_mar[d[3]],df_mar[d[4]]))
df_mar['temp'] = df_mar['temp'].apply(lambda x: list(x))
df_mar = df_mar.set_index(d[2])
idx_mar = df_mar.index.repeat(df_mar['temp'].str.len(), axis = None)
rep_mar = pd.DataFrame({'Impressions':np.concatenate(df_mar['temp'].values)},index = idx_mar)
rep_mar['Type'] = six
df_mar = df_mar.join(rep_mar, how='left')
df_mar.drop(columns = ['Mono Pages','Colour Pages','temp'], inplace = True)

# Apr
df_apr['temp'] = list(zip(df_apr[d[3]],df_apr[d[4]]))
df_apr['temp'] = df_apr['temp'].apply(lambda x: list(x))
df_apr = df_apr.set_index(d[2])
idx_apr = df_apr.index.repeat(df_apr['temp'].str.len(), axis = None)
rep_apr = pd.DataFrame({'Impressions':np.concatenate(df_apr['temp'].values)},index = idx_apr)
rep_apr['Type'] = five
df_apr = df_apr.join(rep_apr, how='left')
df_apr.drop(columns = ['Mono Pages','Colour Pages','temp'], inplace = True)

# May
df_may['temp'] = list(zip(df_may[d[3]],df_may[d[4]]))
df_may['temp'] = df_may['temp'].apply(lambda x: list(x))
df_may = df_may.set_index(d[2])
idx_may = df_may.index.repeat(df_may['temp'].str.len(), axis = None)
rep_may = pd.DataFrame({'Impressions':np.concatenate(df_may['temp'].values)},index = idx_may)
rep_may['Type'] = five
df_may = df_may.join(rep_may, how='left')
df_may.drop(columns = ['Mono Pages','Colour Pages','temp'], inplace = True)
```

```python
df_jan = df_jan.reset_index()
df_feb = df_feb.reset_index()
df_mar = df_mar.reset_index()
df_apr = df_apr.reset_index()
df_may = df_may.reset_index()
```

```python
df_all = rc.mergeData(d1 = df_jan, d2 = df_feb, d3 = df_mar, d4 = df_apr, d5 = df_may).reset_index()
df_all.drop(columns = 'index', inplace = True)
```

```python
df_loc = df_all.groupby(['Lokasi','Periode'])['Impressions'].sum()
df_eq = df_all.groupby(['EQ','Periode'])['Impressions'].sum()
df_eq_sum = df_all.groupby(['EQ','Lokasi'])['Impressions'].sum().sort_values(ascending = False)

dl = df_loc.reset_index()
de = df_eq.reset_index()

df_all.to_excel("D:\\Data Shelter\\Kerjaan\\Astragraphia\\BO1\\Pelindo\\Zweite Entwicklung\ILCS_6_mo.xlsx")
```
